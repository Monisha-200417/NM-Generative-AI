```python
from transformers import pipeline

# Load pre-trained zero-shot classification model
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Labels for spam detection
labels = ["Spam", "Not Spam"]

def predict_spam(text):
    result = classifier(text, labels)
    prediction = result["labels"][0]
    score = result["scores"][0]
    return f"\nPrediction: {prediction} ({score * 100:.2f}% confidence)\n"

def main():
    print("📩 Spam Detector (powered by Hugging Face Transformers)")
    print("Type 'exit' to quit.\n")
    while True:
        user_input = input("Enter a message: ")
        if user_input.lower() == "exit":
            break
        output = predict_spam(user_input)
        print(output)

if __name__ == "__main__":
    main()

```
